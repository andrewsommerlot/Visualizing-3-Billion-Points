**Background**

The Chesapeakebay Bay Model is used to guide decisions governing a multi-agency environmental effort in the Eastern United States. Excess nutrients, especially nitrogen and phosphorus from all kinds of actions make their way into the Chesapeake Bay. One main indicator of bay health is dissolved oxygen (DO). When DO levels become too low, really nasty things can happen like agal blooms that can make the water poisonous to people! This cuases all kinds of damage to our health, our property values, and if you depend on the Bay economy for a job, then your income too. Problems like these are very hard to fix after the fact, so we try to keep them from happening. One way that modeling helps this process is to study different scenarios of human actions and how they will impact the Bay for the better or worse. 

Implementation plans are big lists of actions that different states will try to take to reduce the impact we have on the Bay. We model the impact of the implementation plan, and other scenarios for comparison. In this case, I have three scenarios, the implementation plan and 2 hypothetical situations: 1) Adding a million pounds of nitrogen flow from the largest drainage area in the watershed into the Bay, and 2) Adding half a million pounds of phosphorus from the same watershed, which will also flow into the Bay. We run the model and look at the estimated changes in dissolved oxygen in the Bay. Easy peasy...

One big challenge we face is the shear number of estiamtes the model we build produces and how to best analyse them to get the most valuable information possible. This is not an academic excercise as we are informing the spending a large budget. The model makes hourly esitimates of DO for thousands of locations in a 3D grid covering the entire volumne of the Bay. One place our analysis could use improvement is looking at the dissolved oxygen estiamtes in open water (near the surface) areas of the Bay. We are better at modeling the deep part of the main channel of the Bay and analyzing its health, but open water tends to be trickier. One reason is we dont usually look at the whole picture since just in open water, we have billions of estiamtes across time, space, and our scenarios. When we aggregate for analysis we lose information.

In this notebook, I'm going to demonstrate the begining of an exploritory process we are using to study how human actions influence Bay health to help inform decisions with dollar signs attached. I wanted to start by trying to visualize every single one of the 3 billion points in this 150 gigabyte data set of open water dissolved oxygen estimates. 

---

**Data info:**

Houry estiamtes of DO across an entire gridded surface of the chesapeake Bay. The time range is ten cycles of the time window June 1st to September 30th from 1990 to 1999. There are 3 distict groups of these model estimates, each with ~1e9 esimates:
* the implementation plan 
* Adding 1 million pounds of nitrogen
* adding half a million pounds of phosphorus.

## Colors For Heat Maps

Things like occusion, and binning inconsistancies are address in datashader. The result of many points all plotted in the same space will create a heat map with a number of different transfer functions. For this notebook, I used the "eq_hist" or histogram equalization because it doesnt make any assumptions about the shape of the distribution. That seems like a good place to start. There are a others preloaded in datashader, including log which is also used alot.

## Making the Plots

I used the datashader points method on a datashader canvas to create the plots. Writing a function uses less ram, which turned out to be nessesary for me (16 GB on my laptop). It was dissapointing though, to find I could only make the images about 800 x 200. This is much smaller than what I was plannig on. After reseaching it quite a bit, it looks like its my ram limitation and I could be smarter with my use of it as well. Although this was dissapointing, I wanted to move on and finish it any way. Even if the plot is too small, it is a great demonstration of what can be done even with very humble hardware. Still a bummer though. 

```python
#====================================================================
#
# source: https://anaconda.org/jbednar/uber/notebook
# may have to redo for all dfs to make sure they all fit in here. 
#
#====================================================================

# time code range
x_range = (151, 3561)

# DO range. There are negetive values here!
y_range =(-13.2, 20.5)

def create_image(df, x_range, y_range, w, h, cmap):
    cvs = ds.Canvas(plot_width=w, plot_height=h, x_range = x_range, y_range = y_range)
    agg = cvs.points(df, 'x', 'y')
    return tf.shade(agg, cmap = cmap, how='eq_hist')
```

```python
#====================================================================
#
# create the first image 
# This takes a little while
#
#====================================================================


# 600 x 200 is the size I assumed for making the weave script below. So, if you change this 
#  you'll have to tune the weave script
d1_im = create_image(d1, x_range, y_range, 600, 200)
```

```python
#====================================================================
#
# export the first image to disk and print out 
#
#====================================================================

# export the image
export_image(tf.set_background(d1_im, 'black'),'pics/susp')
```

```python
#==================================================================
#
# create the second image 
# Right now, these have to be done one at a time in unique notebook launches. 
# Dask still needs a lot of ram the way I'm using it here. 
#
#==================================================================

# run the create image function. 
d2_im = create_image(d2, x_range, y_range, 600, 200, cmap = cool)
```

```python
#====================================================================
#
# export the second image 
#
#====================================================================

# export image, also prints to console
export_image(tf.set_background(d2_im, 'black'),'pics/wip')
```

```python
#==================================================================
#
# create the third image 
#
#==================================================================

# export image, also prints to console
d3_im = create_image(d3, x_range, y_range, 600, 200, cmap = neon)
```

```python
#====================================================================
#
# export the third image 
#
#====================================================================

# export image, also prints to console
export_image(tf.set_background(d3_im, 'black'),'pics/susn')
```

## Weave the Plots Together with Image Processing

Here, I wrote a quick script to slice all the data parts out of the images and weave them together into one plot. I found I could not do this with datashader, as its kind of a weird idea. The empty space in the plots gave me the idea to ignore the x acess placement and just put them all up next to each other. Kind of like holding up to x-rays to the light to compare them. 

There is an interesting feature in the data that can be seen as horizontal lines going through the bottom quarter or so of all the swatches of data. This is the zero line, and there are negetive values! Negetive DO? Not possible. True, but these are model estimates and the way this model works mass has to be presevered through the run, and do to time lags, sometimes the model tracks a negetive DO to preseve mass of neutrients as they move through the Bay. 

We knew this was happening, but considered it a trivial eccentricity of the model. When you see it as one of the most obvious patterns across all 150 gigabytes of the data, it makes you look again. We had never before counted how many times zeros exisited in the dataset. We did different cell groups individually, but the pattern was never as obvious until the whole set was plotted. 

```python
#==============================================================================================
#
#  Slice the images and weave them back together one slide from one image next to eachother
#
#  using cv2 
#
#===========================================================================================

# read in the output images from datashader separately. 
img_cool = mpimg.imread('pics/wip.png')
img_hot = mpimg.imread('pics/susp.png')
img_neon = mpimg.imread('pics/susn.png')

# hand picked number through trial and error. Depends entirely on x size of image. 
# here I was plotting a image 600 pixes wide 

# !!!!! this changes everything, tune it by hand for pictures with new widths
plus = 64

# define empty lists 
cool_slices = []
hot_slices = []
neon_slices = []

# not the best form, but who cares for now
for i in range(0,10):
    cool_slice = img_cool[:,0+plus*i:25+plus*i,:]
    hot_slice = img_hot[:,0+plus*i:25+plus*i,:]
    neon_slice = img_neon[:,0+plus*i:25+plus*i,:]
    cool_slices.append(cool_slice)
    hot_slices.append(hot_slice)
    neon_slices.append(neon_slice)


# correct for the one missing width pixel column on the last slice, annoying
for slices in [cool_slices, hot_slices, neon_slices]:
    replacement = np.zeros([200,25,4])
    replacement[0:200, 0:24, 0:4] = slices[9]
    slices[9] = replacement


# define new image to weave everything together in
new_image = np.zeros([200, 895, 4])

# place each slice into the new image to create side by side comparison
for i in range(0, 10):
    new_image[:,10+0+(plus+25)*i:10+25+(plus+25)*i,:] = cool_slices[i]
    new_image[:,10+25+(plus+25)*i:10+25+25+(plus+25)*i,:] = hot_slices[i]
    new_image[:,10+50+(plus+25)*i:10+50+25+(plus+25)*i,:] = neon_slices[i]

    
# opacity for all black backgorund on new image
new_image[:,:,3] = 1

# define figure size 
plt.figure(figsize=(18,18))
plt.imshow(new_image)

# also save a png to disk 
plt.imsave('pics/weave.png', new_image)
```
